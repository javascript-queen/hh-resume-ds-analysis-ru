1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:

- Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
- Неоконченное высшее образование 2000 Балтийская государственная академия рыбопромыслового флота…
  Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

_Совет: проверяйте полученные категории, например, с помощью метода unique()_


```python
# Сколько соискателей имеет средний уровень образования (школьное образование)?

# Функция для определения категорий образования
def categorize_education(row):
    # Получаем первые три слова
    words = row.split()[:3]
    # Объединяем их обратно в строку
    first_three_words = ' '.join(words)
    # Проверка на вход слов в выборку
    if 'Высшее' in first_three_words:
        return 'высшее'
    elif 'Неоконченное высшее' in first_three_words:
        return 'неоконченное высшее'
    elif 'Среднее специальное' in first_three_words:
        return 'среднее специальное'
    elif 'Среднее' in first_three_words:
        return 'среднее'


# Создаем новый столбец с категориями образования
df['Образование'] = df['Образование и ВУЗ'].apply(categorize_education)

# Сколько всего получилось грууп (должно быть 4)
print(df['Образование'].value_counts())

# Подсчитываем количество соискателей среднего уровня образования для задания на платформе
count_middle_education = df[df['Образование'] == 'среднее'].shape[0]
print(
    f"Количество соискателей со средним образованием: {count_middle_education}")
```

  Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:

- Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина.
- Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

```python
# Создаем столбец 'Пол'
df['Пол'] = df['Пол, возраст'].apply(lambda x: x.split(',')[0].strip())

# Функция для разделения по половому признаку


def categorize_gender(row):
    # Проверка, содержит ли строка информацию о поле 'Мужчина'
    if 'Мужчина' in row:
        return 'М'  # Возвращает 'М' в случае, если строка содержит информацию о мужском поле
    # Проверка, содержит ли строка информацию о поле 'Женщина'
    elif 'Женщина' in row:
        return 'Ж'  # Возвращает 'Ж' в случае, если строка содержит информацию о женском поле


# Применяем функцию к столбцу ['Пол]
df['Пол'] = df['Пол'].apply(categorize_gender)

# Создаем столбец 'Возраст'
df['Возраст'] = df['Пол, возраст'].apply(
    lambda x: int(x.split(',')[1].split()[0]))

# Удаляем столбец 'Пол, возраст'
df.drop(columns=['Пол, возраст'], inplace=True)

display(df.head())

# ДОПОЛНИТЕЛЬНЫЕ ВОПРОСЫ

# Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых.

# Подсчет количества женских резюме
female_resume = df[df['Пол'] == 'Ж']['Пол'].count()

# Подсчет общего числа резюме
total_resume = df['Пол'].count()

# Вычисление процента женских резюме
percentage_female = (female_resume / total_resume) * 100

# Ответ
print(f"Процент женских резюме в данных: {percentage_female:.2f}%")

# Чему равен средний возраст соискателей? Ответ округлите до десятых
print(f"Средний возраст соискателей: round(df['Возраст'].mean(), 2)")
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**.

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:

- Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
- Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
- В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
- В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
  - Опыт работы 3 года 2 месяца…
  - Опыт работы 4 года…
  - Опыт работы 11 месяцев…
  - Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def extract_experience(row):
    # # Проверка на пустое значение или значение 'Не указано'
    if pd.isnull(row) or row == 'Не указано':
        # Возвращает NaN, если строка пустая или содержит 'Не указано'
        return np.nan
    # Списки ключевых слов для определения информации о годах и месяцах опыта
    year_keywords = ['год', 'года', 'лет']
    month_keywords = ['месяц', 'месяца', 'месяцев']

    # Получение первых 6 слов из строки опыта
    words = str(row).split()[:6]
    months = 0  # Переменная для хранения количества месяцев
    years = 0  # Переменная для хранения количества лет

    # enumerate позволяет получить индекс каждого элемента списка words
    for i, word in enumerate(words):
        if word in year_keywords:
            # Получение количества лет из строки
            years = int(words[i - 1])
        elif word in month_keywords:
            # Получение количества месяцев из строки
            months = int(words[i - 1])

    # Подсчет общего количества месяцев опыта
    total_months = years * 12 + months
    # Возвращаем общее количество месяцев опыта
    return total_months


# Применяем функцию к столбцу 'Опыт работы' и создаем новый столбец 'Опыт работы (месяц)'
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(extract_experience)

# Удаление столбца 'Опыт работы'
df.drop('Опыт работы', axis=1, inplace=True)
print(df['Опыт работы (месяц)'])

# ДОПОЛНИТЕЛЬНЫЕ ВОПРОСЫ

# Находим медианный опыт работы
median_experience = df['Опыт работы (месяц)'].median()
print(f"Медианный опыт работы (в месяцах): {median_experience}")
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

- Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

  Список городов-миллионников:

  <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
  </code>
  Инфорация о метро, рядом с которым проживает соискатель нас не интересует.

- Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к переезду в признаке "Город, переезд, командировки". Например:
  - … , готов к переезду , …

    Нас интересует только сам факт возможности или желания переезда.
- Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к командировкам в признаке "Город, переезд, командировки". Например:
  - … , готов к командировкам , …

    Нас интересует только сам факт готовности к командировке.
    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

_Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе._


```python
# Функция для категоризации городов на основе данных о месте проживания и перемещениях.
def categorize_cities(row):
    # Список крупнейших городов России
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
                      'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']

    # Получение первого слова из строки о городе, переездах, командировках
    row = row.split(' ')[:1]
    for word in row:
        # Проверка на наличие слова 'Москва'
        if 'Москва' in word:
            return 'Москва'
        # Проверка на наличие слова 'Санкт-Петербург'
        elif 'Санкт-Петербург' in word:
            return 'Санкт-Петербург'
        # Проверка на наличие городов-миллионников
        elif word in million_cities:
            return 'город-миллионник'
        # Возвращает 'другие' в случае отсутствия совпадений
        else:
            return 'другие'

# Применение функции categorize_cities к столбцу 'Город, переезд, командировки' в датафрейме df
df['Город'] = df['Город, переезд, командировки'].apply(categorize_cities)
display(df['Город'])

# ДОПОЛНИТЕЛЬНЫЕ ВОПРОСЫ

# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого
spb = df[df['Город'] == 'Санкт-Петербург']['Город'].count()
total = df['Город'].count()
percentage_spb = (spb / total) * 100
# Вывод округленного процента
print(f' Сколько процентов соискателей живут в Санкт-Петербурге: {percentage_spb}')

# Функция для извлечения информации о готовности к командировкам
def extract_trips(row):
    if ('командировка' in row):
        if ('не готов к командировкам' in row) or ('не готова к командировкам' in row):
            return False
        else:
            return True
    else:
        return False

# Применение функции extract_trips к столбцу 'Город, переезд, командировки' в датафрейме df
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(
    extract_trips)
display(df['Готовность к командировкам'].value_counts())

# Функция для извлечения информации о готовности к переезду
def extract_move(row):
    if ('не готов к переезду' in row) or ('не готова к переезду' in row):
        return False
    elif 'хочу переехать' in row:
        return True
    else:
        return True

# Применение функции extract_move к столбцу 'Город, переезд, командировки' в датафрейме df
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(
    extract_move)
display(df['Готовность к переезду'])
display(df['Готовность к переезду'].value_counts())

# Расчет процента соискателей, готовых и к переездам, и к командировкам
ready = ((df['Готовность к командировкам'] == True) &
          (df['Готовность к переезду'] == True)).sum()
total_candidates = len(df)
percent_ready_for_both = (ready / total_candidates) * 100
percent_ready_for_both_rounded = round(percent_ready_for_both)
print(
    f"Процент соискателей, готовых к переездам и командировкам: {percent_ready_for_both_rounded}%")
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.


Итак, давайте обсудим возможный алгоритм преобразования:

1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Читаем новую таблицу с валютами
currency = pd.read_csv('./ExchangeRates.csv')
display(currency)

# Преобразования с датами
df['Обновление резюме'] = df['Обновление резюме'].apply(lambda x: x.split()[0])
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'])
df['Обновление резюме'] = df['Обновление резюме'].dt.strftime('%d/%m/%Y')

currency['date'] = pd.to_datetime(currency['date'])
currency['date'] = currency['date'].dt.strftime('%d/%m/%Y')

# Создаем новый столбец 'Сумма ЗП' и 'Валюта'
df[['Сумма ЗП', 'Валюта']] = df['ЗП'].str.extract(r'(\d+)\s+(\D+)')

# Таблица соответствия валюты к ISO
currency_ISO = {
    'руб.': 'RUB',
    'USD': 'USD',
    'грн.': 'UAH',
    'EUR': 'EUR',
    'бел.руб.': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}

df['Валюта_ISO'] = df['Валюта'].map(currency_ISO)

# Объединение таблиц по столбцам с датой и валютой
merged_data = pd.merge(df, currency, left_on=[
                       'Обновление резюме', 'Валюта_ISO'], right_on=['date', 'currency'], how='left')

# Заполнение пропущенных значений close и proportion для рубля
merged_data.loc[merged_data['Валюта_ISO']
                == 'RUB', ['close', 'proportion']] = 1

# Умножение и деление для получения ЗП в рублях
merged_data['ЗП (руб)'] = merged_data['Сумма ЗП'].astype(
    float) * merged_data['close'] / merged_data['proportion']

# Удаление промежуточных столбцов
merged_data.drop(['Сумма ЗП', 'Валюта', 'Валюта_ISO', 'date', 'time',
                 'currency', 'close', 'vol', 'proportion', 'per'], axis=1, inplace=True)

# Рассчитать медианную заработную плату в рублях
median_salary_rubles = merged_data['ЗП (руб)'].median()

# Конвертировать медиану в тысячи и округлить до ближайшего целого
median_salary_thousands = round(median_salary_rubles / 1000)

# Результат
display(merged_data)
print(f"Медианная заработная плата: {median_salary_thousands} тыс. рублей")

df = merged_data.copy()
```

**2. Анализ булевых значений при помощи метода Крамера.**

Для булевых значений лучше проверить Коэффициент φ Крамера:

1. Подсчитать таблицу сопряженности для двух булевых признаков. Таблица сопряженности показывает, сколько раз происходит комбинация значений признаков.

2. Рассчитать хи-квадрат (χ²) — статистический тест для проверки наличия зависимости между двумя категориальными переменными.

3. Используя значение хи-квадрат, рассчитать коэффициент φ Крамера по формуле:

$$
ϕ_c = \sqrt \frac{x^2}{n⋅min⁡((r−1),(c−1))}
$$
​

где:
ϕc​ - коэффициент φ Крамера,

χ2 - значение хи-квадрат,

n - общее количество наблюдений,

r - количество строк в таблице сопряженности,

c - количество столбцов в таблице сопряженности.

Я предполагаю, что есть возможная корреляция сежду теми, кто готов к переезду и теми, кто готов к командировкам. Попробуем использовать метод коэффициента Крамера. Для этого мы сначала напишем формулу нахождения кэффициента, потом построим новую матрицу сопряжённости для признаков.

```python
# Создаем таблицу сопряженности для двух столбцов
observed = pd.crosstab(numerical_df['Готовность к командировкам'], numerical_df['Готовность к переезду'])

# Переводим таблицу сопряженности в формат numpy array
observed_array = observed.to_numpy()

# Рассчитываем количество строк и столбцов в таблице
n_rows, n_cols = observed_array.shape

# Рассчитываем суммы по строкам и столбцам
row_sum = observed.sum(axis=1)
col_sum = observed.sum(axis=0)
total_sum = observed.sum().sum()

# Рассчитываем ожидаемые частоты
expected = np.outer(row_sum, col_sum) / total_sum

# Рассчитываем коэффициент хи-квадрат
chi2 = np.sum((observed_array - expected)**2 / expected)

# Рассчитываем коэффициент Крамера
kramer_coefficient = np.sqrt(chi2 / (total_sum * (min(n_rows, n_cols) - 1)))

print("Коэффициент Крамера:", kramer_coefficient)

# Построение графика (необязательно)
plt.figure(figsize=(6, 4))
plt.imshow(observed, cmap='BuPu', interpolation='nearest')
plt.colorbar()
plt.xlabel('Готовность к переезду')
plt.ylabel('Готовность к командировкам')
plt.title('Таблица сопряженности')
plt.show()

display(observed)
```

```python
# Создаем график распределения возраста в логарифмическом масштабе
fig, ax = plt.subplots(1, 1, figsize=(8, 4))

# Преобразуем возраст в логарифмическую шкалу
log_age = np.log(df['Возраст'] + 1)

# Строим гистограмму
histplot = sns.histplot(log_age, bins=30, ax=ax, color='#9370DB')

# Добавляем линию для среднего значения
histplot.axvline(log_age.mean(), color='m', lw=2)

# Добавляем линии для границ интервала 3σ и 4σ
histplot.axvline(log_age.mean() + 4 * log_age.std(), color='m', ls='--', lw=2)
histplot.axvline(log_age.mean() - 3 * log_age.std(), color='m', ls='--', lw=2)

# Устанавливаем заголовок
histplot.set_title('Распределение возраста в логарифмическом масштабе')
plt.ylabel('Количество соискателей')

# Функция для определения выбросов по z-отклонению и очистки данных
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Находим выбросы и чистим данные
outliers, cleaned_data = outliers_z_score_mod(
    df, 'Возраст', left=3,  right=4, log_scale=True)

# Выводим количество выбросов
print(
    f"Выбросы по методу z-отклонения (>4σ) в правую сторону: {outliers.shape[0]}")

# Определение возраста соискателей, попавших в категорию выбросов
print('Возраст соискателей, попавших в категорию выбросов')
outliers_age = outliers['Возраст']
display(outliers_age)
```
